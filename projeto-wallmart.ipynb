#Lojas com melhores desempenhos

```python
biggest_sales_store = df.groupby(['Date', 'Store']).sum().sort_values('Weekly_Sales',ascending=False).round()
biggest_sales_store.head()
```

```python
max = df.groupby('Store')['Weekly_Sales'].max().tolist()
min = df.groupby('Store')['Weekly_Sales'].min().tolist()
mean_efetivos = df[df['Holiday_Flag']==0].groupby('Store')['Weekly_Sales'].mean().tolist()
std = df.groupby('Store')['Weekly_Sales'].std().tolist()
mean_feriados = df[df['Holiday_Flag']==1].groupby('Store').Weekly_Sales.mean().tolist()
store = df.Store.unique()
dicionario = {
    'store': store,
    'max' :max,
    'min': min,
    'mean': mean_efetivos,
    'std': std,
    'Vendas_feriados': mean_feriados
    }
analitcs = pd.DataFrame(dicionario)

w=0.2

plt.figure(figsize=(25,10))
bar1 = plt.bar(store[35:45], analitcs.sort_values(by='mean')['max'].tail(10), width=w)
bar2 = plt.bar(store[35:45]+w, analitcs.sort_values(by='mean')['min'].tail(10), width=w)
bar3 = plt.bar(store[35:45]+w*2, analitcs.sort_values(by='mean')['mean'].tail(10), width=w)
bar4 = plt.bar(store[35:45]+w*3, analitcs.sort_values(by='mean')['std'].tail(10), width=w)
bar5 = plt.bar(store[35:45]+w*4, analitcs.sort_values(by='mean')['Vendas_feriados'].tail(10), width=w)

plt.xticks(store[35:45]+w*1.5,analitcs.sort_values(by='mean').store.tail(10))
plt.legend([bar1, bar2, bar3, bar4, bar5],['Máx','Mín','Média - Efetivos','Desvio padrão','Média-Feriados'], loc=2, fontsize=20)
plt.title('Vendas máx, mín, médias, desvio padrão e médias nos feriados de todas as Lojas - 05/02/10 até 01/11/12', fontsize=20)
plt.show()
```
